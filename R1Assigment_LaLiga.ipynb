```python
#4) WWrite a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)
```

```python
def Goal_diff_count():
    laliga['GoalsFor']=laliga['GoalsFor'].apply(pd.to_numeric)
    laliga['GoalsAgainst']=laliga['GoalsAgainst'].apply(pd.to_numeric)
    laliga['GoalDiff']=laliga['GoalsFor']-laliga['GoalsAgainst'] # Instead of variable adding one more column in dataframe and displaying it
    pd.options.display.max_columns = None
    display(laliga.loc[laliga['GoalDiff']==laliga['GoalDiff'].max()])
    display(laliga.loc[laliga['GoalDiff']==laliga['GoalDiff'].min()])
```

```python
Goal_diff_count()
```

```python
#5)Create a new column with the name “Winning Percent” and append it to the data set (5 points)
```

```python
# Print the top 5 teams which have the highest Winning percentage
```

```python
def Winning_Percentage():
    laliga['GamesWon']=laliga['GamesWon'].apply(pd.to_numeric)
    laliga['GamesPlayed']=laliga['GamesPlayed'].apply(pd.to_numeric)
    try:
        laliga['Percentage of Winning']=(laliga['GamesWon']/laliga['GamesPlayed'])*100
    except ZeroDivisionError:
        laliga['Percentage of Winning']=0
        laliga.fillna(0, inplace=True)
    pd.options.display.max_columns = None
    pd.options.display.max_rows = None
    laliga.sort_values('Percentage of Winning',ascending=0)
    display(laliga.head(5))
Winning_Percentage()  
```
